```python
#Dealing with the NaN
data = data.astype(object).where(pd.notnull(data),None)
data.isna().any()
```

```python
#Missing values per column
def num_missing(x):
  return sum(x.isnull())
#Applying per column:
print ("Missing values per column:")
print (data.apply(num_missing, axis=0)) #axis=0 defines that function is to be applied on each column

#Applying per row:
print ("Missing values per row:")
print (data.apply(num_missing, axis=0).head()) #axis=1 defines that function is to be applied on each row





```

```python
def convert_currency_amountmin(row):
    if row['prices.currency'] == 'CAD':
        newval = row['prices.amountmin'] * .73 
    return newval  

data['prices.currency'] = data.apply(convert_currency_amountmin, axis = 1)
```

```python
# Transform different currency to USD

for row in data.itertuples():
    if row  == 'CAD':
        data['prices.amountmin'][row.Index] *= 0.73
        data['prices.amountmax'][row.Index] *= 0.73
        data['prices.amountavg'][row.Index] *= 0.73
    elif row == 'EUR':
        data['prices.amountmin'][row.Index] *= 1.1
        data['prices.amountmax'][row.Index] *= 1.1
        data['prices.amountavg'][row.Index] *= 1.1
    elif row == 'AUD':
        data['prices.amountmin'][row.Index] *= 0.75
        data['prices.amountmax'][row.Index] *= 0.75
        data['prices.amountavg'][row.Index] *= 0.75
    elif row == 'GPB':
        data['prices.amountmin'][row.Index] *= 1.3
        data['prices.amountmax'][row.Index] *= 1.3
        data['prices.amountavg'][row.Index] *= 1.3 
```
